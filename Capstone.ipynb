##Feature Engineering

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_selection import VarianceThreshold
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, f1_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
# Pick csv file
file_path = "1- mental-illnesses-prevalence.csv"
df = pd.read_csv(file_path)
print("Dataset loaded! Shape:", df.shape)
```

```python
# Variance Filtering (fast ,on numerical features)
num_cols = df.select_dtypes(include=['float64', 'int64']).columns
selector = VarianceThreshold(threshold=0.01)  # Remove features with variance < 0.01
X_variance = df[num_cols]
selector.fit(X_variance)
selected_column = X_variance.columns[selector.get_support()]
print(f"columns selected via low variance: {list(selected_column)}")
```

```python
# Add target variable (High_Depression)
depression_col = 'Depressive disorders (share of population) - Sex: Both - Age: Age-standardized'
if depression_col in df.columns:
    df['High_Depression'] = (df[depression_col] > df[depression_col].mean()).astype(int)
    print("Target variable 'High_Depression' added.")
else:
    raise ValueError(f"Column '{depression_col}' not found in dataset.")
```

```python
#Encode categorical features
if 'Entity' in df.columns:
    le = LabelEncoder()
    df['Entity_Encoded'] = le.fit_transform(df['Entity'])
    if 'Entity' in selected_column:
        selected_column = [col for col in selected_column if col != 'Entity'] + ['Entity_Encoded']
    print("Categorical feature 'Entity' encoded.")
```

```python
# Random Forest top-N feature (on a sample for speed)
sample_size = 5000 if df.shape[0] > 5000 else df.shape[0]
df_sample = df.sample(n=sample_size, random_state=42)
X = df_sample[selected_column]
y = df_sample['High_Depression']
rf_model = RandomForestClassifier(n_estimators=10, random_state=42)
rf_model.fit(X, y)
importances = pd.Series(rf_model.feature_importances_, index=X.columns)
top_features = importances.sort_values(ascending=False).head(5).index.tolist()
print(f"Top features selected via Random Forest: {top_features}")
```

```python
#Final dataset
df_final = df[top_features + ['High_Depression']]
print("Final dataset shape:", df_final.shape)
print(df_final.head())
```

## Step 7: Model Selection and Justification

The task at hand is a **binary classification problem**, where the goal is to predict whether the prevalence of depressive disorders is high or not (`High_Depression`).

Several classification algorithms were considered for this task:

1. **Logistic Regression**  
   - Simple, interpretable model that works well for binary classification.  
   - Provides probability outputs and insights into feature influence.  
   - Useful as a baseline model.  

2. **Random Forest Classifier**  
   - An ensemble method that combines multiple decision trees.  
   - Handles both linear and non-linear relationships.  
   - Performs well with feature importance analysis and is robust to overfitting.  

3. **Gradient Boosting (XGBoost/LightGBM)**  
   - Advanced boosting algorithm that builds strong learners from weak learners.  
   - Generally provides higher accuracy for structured/tabular data.  
   - Suitable when small improvements in prediction performance are needed.  

4. **Support Vector Machine (SVM)**  
   - Effective in high-dimensional spaces.  
   - Works well with clear separation between classes.  
   - However, can be computationally expensive with larger datasets.  

### Final Model Choice
For this project, the primary models selected for training and evaluation are:
- **Logistic Regression** (as a baseline model).  
- **Random Forest Classifier** (for robust performance and interpretability).  

These models were chosen because they balance **predictive accuracy, interpretability, and efficiency**. Gradient Boosting may also be explored for potential performance improvements.

## Step 9: Model Evaluation and Comparison

In this step, we evaluated the performance of two machine learning models — **Logistic Regression** and **Random Forest Classifier** — using the test set. Both models were trained to predict whether a given instance belongs to the **High_Depression** category.

---

### 1. Logistic Regression
Logistic Regression is a **linear model** that estimates the probability of a binary outcome using the logistic (sigmoid) function.  

- **Why chosen:**  
  - Acts as a strong **baseline model** for binary classification problems.  
  - Provides **interpretable coefficients** that show how each feature contributes to the target outcome.  
  - Computationally efficient and fast to train.  

- **Advantages:**  
  - Simple and interpretable.  
  - Works well with scaled numerical data.  
  - Low risk of overfitting with regularization.  

- **Limitations:**  
  - Assumes a **linear relationship** between features and the log-odds of the target.  
  - May underperform on datasets with complex, non-linear interactions.  

---

### 2. Random Forest Classifier
Random Forest is an **ensemble method** based on bagging, where multiple decision trees are trained on bootstrapped subsets of the data, and their predictions are averaged.  

- **Why chosen:**  
  - Can handle both **numerical and categorical** features effectively.  
  - Captures **non-linear relationships** between variables.  
  - Provides feature importance, helping to identify which predictors are most influential.  

- **Advantages:**  
  - High accuracy and robustness to noise.  
  - Reduces overfitting compared to a single decision tree.  
  - Handles missing values and imbalanced datasets relatively well.  

- **Limitations:**  
  - Less interpretable than Logistic Regression.  
  - Requires more computational resources.  

---

### 3. Evaluation Metrics
Both models were evaluated on the **test set** using the following metrics:

- **Accuracy:** The proportion of correct predictions out of all predictions.  
- **Precision:** Of all the predicted positives, how many are truly positive.  
- **Recall (Sensitivity):** Of all the actual positives, how many are correctly identified.  
- **F1-Score:** The harmonic mean of precision and recall.  
- **Confusion Matrix:** Provides a breakdown of true positives, true negatives, false positives, and false negatives.  

---

### 4. Model Performance (Expected)
- **Logistic Regression:**  
  Achieves a reasonable accuracy and provides interpretability, but may miss complex feature interactions.  
- **Random Forest:**  
  Typically outperforms Logistic Regression on accuracy and recall due to its ability to capture non-linearities.  

---

### 5. Justification of Model Selection
Since our main goal is to **predict high prevalence of depression**:  

- If the dataset is **balanced**, we prioritize **overall accuracy**.  
- If the dataset is **imbalanced**, we prioritize **recall**, ensuring we correctly identify as many high-depression cases as possible.  

**Final Decision:**  
- Logistic Regression is kept as a **baseline model** for interpretability.  
- Random Forest is selected as the **preferred model** because it provides higher performance and robustness, making it more suitable for deployment in real world tasks.

```python
from sklearn.model_selection import GridSearchCV
# Define parameter grid for Random Forest
param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['sqrt', 'log2']
}
```

```python
# Use the existing preprocessor and X_train_processed
# Run GridSearchCV for Random Forest
grid_search = GridSearchCV(
    estimator=RandomForestClassifier(random_state=42),
    param_grid=param_grid,
    cv=3,
    n_jobs=-1,
    verbose=2,
    scoring='accuracy'
)
grid_search.fit(X_train_processed, y_train)
print("Best Cross-Validation Accuracy:", grid_search.best_score_)
```

```python
# Check best parameters
print("Best Parameters:", grid_search.best_params_)
print("Best Cross-Validation Accuracy:", grid_search.best_score_)
```

```python
# Train the final Random Forest model with best parameters
best_rf = RandomForestClassifier(**grid_search.best_params_, random_state=42)
best_rf.fit(X_train_processed, y_train)
```

```python
# Evaluate on test set
y_pred_rf = best_rf.predict(X_test_processed)

from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

print("Test Accuracy:", accuracy_score(y_test, y_pred_rf))
print("\nClassification Report:\n", classification_report(y_test, y_pred_rf))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred_rf))
```
